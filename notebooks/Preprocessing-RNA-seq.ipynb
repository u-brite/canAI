```python
import pandas as pd
data=pd.read_csv('TCGA-PRAD.htseq_counts.tsv',sep='\t')
```

```python
data1 = data.dropna(thresh=(0.3*(len(data.columns)))) #30% NA
data2 = data1.T
```

```python
data2.head()
```

```python
new_header = data2.iloc[0] 
data2 = data2[1:] 
data2.columns = new_header
```

```python
data2.head()
```

```python
sample = data2.var(ddof=0)
```

```python
sample
```

```python
Top_1000 = sample.sort_values(ascending=False).head(1000)
```

```python
Top_1000.head()
```

```python
Top_1000_1 = Top_1000.to_frame()
```

```python
Top_1000_1.head()
```

```python
Top_1000_1.index.name = 'newhead'
Top_1000_1.reset_index(inplace=True)
```

```python
first_column_values = Top_1000_1.iloc[:, 0].values
```

```python
first_column_values
```

```python
Top_1000_data = data2[first_column_values]
Top_1000_data
```

```python
writer = pd.ExcelWriter('RNA-seq-Top1000.xlsx')
Top_1000_data.T.to_excel(writer,'Result',index =True)
writer.save()
```
